```python
def export_legend(ax, filename="Legend.svg"):
    fig2 = plt.figure(figsize=(8,1))
    ax2 = fig2.add_subplot()
    ax2.axis('off')

    legend = ax2.legend(*ax.get_legend_handles_labels(), frameon=False, loc='lower center', fontsize=font_mid, ncol=4,)

    fig  = legend.figure

    fig.canvas.draw()
    bbox  = legend.get_window_extent().transformed(fig.dpi_scale_trans.inverted())
    fig.savefig(filename, dpi=100, bbox_inches=bbox)
```

```python
adv = [76.2,54.36,52.77,52.2,52.09]
nat = [55.73,0.8,0.02,0,0]

x_overall = [1,5,10,20,50]

fig, ax1 = plt.subplots(figsize=(8,6))  # base fig
# fig, ax1 = plt.subplots(figsize=(10,6))  # zoom-in fig

ax1.plot(x_overall, adv, c=color1, marker='o', markersize=3*lw, label=r"$\bf{Robust\ Model}$", lw=lw)
ax1.plot(x_overall, nat, c=color2, marker='o', markersize=3*lw, label=r"$\bf{Standard\ Model}$", lw=lw)

ax1.tick_params(axis = 'both', which = 'major', labelsize = 17)
ax1.tick_params(axis = 'x', which = 'major', labelsize = font_tick_x)
ax1.tick_params(axis = 'y', which = 'major', labelsize = font_tick_y)

# ax1.set_ylim([max(int(min(min(adv),min(nat)))-lower_gap,0), math.ceil(max(max(adv), max(nat)))+upper_gap])
# ax1.set_xlim([0.091,0.92])
# xticks = np.arange(0.2, 1, 0.2)
# ax1.set_xticks(xticks)

ax1.set_xlabel('PGD Attack Iterations', fontsize=font_label, fontweight='bold')
ax1.set_ylabel('Accuracy (%)', fontsize=font_label, fontweight='bold')

ax1.spines['bottom'].set_linewidth(font_board)
ax1.spines['bottom'].set_color('black')
ax1.spines['left'].set_linewidth(font_board)
ax1.spines['left'].set_color('black')
ax1.spines['top'].set_linewidth(font_board)
ax1.spines['top'].set_color('black')
ax1.spines['right'].set_linewidth(font_board)
# ax1.spines['right'].set_linestyle("--")
ax1.spines['right'].set_color('black')

# leg = ax1.legend(fontsize=font_mid, loc='lower right', prop={'size': legend_size}, ncol=1)
#
# leg.get_frame().set_edgecolor("black")
# leg.get_frame().set_linewidth(3)
ax1.grid(axis='both', color='gray', linestyle='-', linewidth=0.3)

export_legend(ax1)
# fig.tight_layout()  # otherwise the right y-label is slightly clipped

plt.savefig("svg/ChangeT.svg", dpi=100)
```
